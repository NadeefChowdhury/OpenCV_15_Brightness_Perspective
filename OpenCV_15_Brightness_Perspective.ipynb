```python
trees = cv.imread('/content/Slanted_Trees_-_geograph.org.uk_-_444114.jpg')
```

```python
cv2_imshow(trees)
```

```python
trees.shape
```

```python
pointsA = np.float32([[0,0], [480, 0], [0,640], [480,640]])
```

```python
pointsB = np.float32([[100,0], [480, 150], [220,640], [480,780]])
```

```python
M2 = cv.getPerspectiveTransform(pointsA, pointsB)
```

```python
warped = cv.warpPerspective(trees, M2, (480, 640))
```

```python
cv2_imshow(warped)
```
